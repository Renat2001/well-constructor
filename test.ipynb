```python
from track import Track

t = Track()
t.kickoff(800)
t.build(2, 60, 0)
t.hold(1000)
t.drop(2, 10, 15) 
t.hold(600)
t.build(2, 80, 10)
t.drop(2, 0, 15)
df = t.finish()
df.plot(x='hd', y='z')
```

```python
df.plot(x='x', y='y')
```
